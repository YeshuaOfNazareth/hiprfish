```python
import skimage
from skimage import io
from matplotlib import pyplot as plt
from matplotlib import cm
from matplotlib_scalebar.scalebar import ScaleBar
from skimage import color
```

```python
def cm_to_inches(x):
    return(x/2.54)
```

```python
sample = '/Users/hao/Documents/Research/Cornell/Projects/HIPRFISH/figures_dropbox/04_02_2019_DSGN0561_fov_2_registered_sum_feature_0_deepcell'
dc_seg = io.imread('{}.tif'.format(sample))
```

```python
colors = cm.get_cmap('tab10')
n = 10
color_list = [colors(i*(1/n)) for i in range(n)]
image_seg_color = color.label2rgb(dc_seg, colors = color_list, bg_color = (0,0,0), bg_label = 0)
```

```python
fig = plt.figure()
fig.set_size_inches(cm_to_inches(5), cm_to_inches(5))
ax = plt.Axes(fig, [0, 0, 1, 1])
fig.add_axes(ax)
ax.imshow(image_seg_color)
scalebar = ScaleBar(0.070, 'um', frameon = True, color = 'white', length_fraction = 0.3, height_fraction = 0.01, 
                    box_color = 'black', box_alpha = 0, location = 1, font_properties = {'size': 8})
plt.gca().add_artist(scalebar)
```

```python
seg_filename = '{}_seg.pdf'.format(sample)
fig.savefig(seg_filename, dpi = 300)
plt.close()
```
